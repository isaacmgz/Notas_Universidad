Las ecuaciones diferenciales ordinarias (EDO) son ecuaciones que involucran funciones y sus derivadas.


<details>
  <summary><b>Glosario:</b></summary>

  __1.__ __`Familia de soluciones`__: Una *familia de soluciones* de una ecuación diferencial es el conjunto de todas las posibles soluciones de la ecuación que se obtienen variando las constantes arbitrarias en la solución general. 


  __2.__ __`Problemas de Valor Inicial (PVI)`__:  Un *Problema de Valor Inicial (PVI)* es una ecuación diferencial junto con condiciones iniciales que especifican el valor de la función desconocida y, en algunos casos, sus derivadas en un punto específico. Estas condiciones iniciales permiten encontrar una solución única de la ecuación diferencial que satisface las condiciones dadas.
  
  __3.__ __`Campo de direcciones`__: El *campo de direcciones* es una herramienta gráfica utilizada para visualizar las soluciones de una ecuación diferencial de primer orden. Representa un campo de vectores en el plano $(x, y)$ donde cada vector indica la dirección de la pendiente de la solución en ese punto.

  __4.__ __`Ecuaciones Diferenciales Homogéneas`__: Una ecuación diferencial es homogénea si puede escribirse en la forma: $L(y)= 0$, donde $L$ es un operador lineal que involucra derivadas de $y$. En otras palabras, todas las partes de la ecuación involucran la función $y$ y sus derivadas, y no hay términos independientes de $y$

  __5.__ __`Ecuaciones Diferenciales No Homogéneas`__: Una ecuación diferencial es no homogénea si incluye un término independiente de $y$ en el lado derecho de la ecuación. Se puede escribir en la forma: $L(y) = g(x)$

  __6.__ __`Ecuación caracteristica`__: La ecuación característica se utiliza para resolver ecuaciones diferenciales lineales homogéneas con coeficientes constantes.

  <details>
   <summary><span style="color: #1073c5;">Ejemplo</span></summary>
 
   Para una ecuación diferencial de segundo orden de la forma:   

   $$y'' + p y' + q y = 0$$   

   la ecuación característica es un polinomio de segundo grado obtenido al asumir una solución de la forma $y = e^{mt}$, donde m es una constante que necesitamos determinar. Siguiendo este supuesto   

   $$ 
   y=e^{mt} \\
   y'=me^{mt} \\
   y''=m^2e^{mt} \\
   $$
  </details>

  __7.__ __`Ecuación Diferencial Exacta`:__ Una ecuación diferencial de primer orden de la forma: $M(x, y) \, dx + N(x, y) \, dy = 0$, es exacta si existe una función $F(x,y)$ tal que:
  $$\frac{\partial F}{\partial x} = M \quad \text{y} \quad \frac{\partial F}{\partial y} = N$$

  Si una ecuación diferencial es exacta, verificamos si se cumple la condición de exactitud:
  $$\frac{\partial M}{\partial y} = \frac{\partial N}{\partial x}$$

  __8__ __`Ecuación Diferencial NO Exacta`:__ Si una ecuación de la forma $M(x, y) \, dx + N(x, y) \, dy = 0$ no cumple con la condición de exactitud, se dice que es no exacta.
</details>

### ☁️ Definiciones y terminología

<details>
  <summary><b>Clasificación de las ecuaciones diferenciales</b></summary>

  ### Tipos de EDO
  
  1. __Por Orden:__
     - __Primer Orden:__ Involucra solo la primera derivada de la función. Ejemplo: $\frac{dy}{dx} = y$.
     - __Segundo Orden:__ Involucra hasta la segunda derivada. Ejemplo: $\frac{d^2y}{dx^2} + \frac{dy}{dx} - y = 0$.
  
  2. __Por Linealidad:__
     - __Lineales:__ La función y sus derivadas aparecen linealmente. Ejemplo: $\frac{dy}{dx} + p(x)y = q(x)$.
     - __No Lineales:__ La función o sus derivadas aparecen de manera no lineal. Ejemplo: $\frac{dy}{dx} = y^2$.

</details>

### ☁️ Casos particulares de Ecuaciones Diferenciales

<details>
  <summary><b>Ecuaciones Diferenciales Homogeneas o No homogeneas</b></summary>

  - __Homogeneas:__ Una ecuación diferencial es homogénea si puede escribirse en la forma: $L(y) = 0$, donde $L$ es un operador lineal que involucra derivadas de $y$ como:  $y'' + p(x)y' + q(x)y = 0$
    - _Solución para Homogeneas:_ se buscan soluciones de la forma $y=y_h$ (solución homogénea) utilizando métodos como la resolución de la ecuación característica en el caso de coeficientes constantes. 
  - __No Homogeneas:__ Una ecuación diferencial es no homogénea si incluye un término independiente de $y$ en el lado derecho de la ecuación. Se puede escribir en la forma: $L(y) = g(x)$ que representa una función tipo: $y''  + p(x)y' + q(x)y = g(x)$
    - _Solución para NO Homogeneas:_ se encuentra una solución particular $y_p$ que satisface la ecuación completa y luego se suma a la solución homogénea $y_h$ para obtener la solución general: y = y_h + y_p

<details>
  <summary><b>Ecuación Diferencial Exacta y No Exacta</b></summary>

  - __Exacta__: Una ecuación diferencial de primer orden de la forma: $M(x, y) \, dx + N(x, y) \, dy = 0$, es exacta si existe una función $F(x,y)$ tal que:
  
  $$\frac{\partial F}{\partial x} = M \quad \text{y} \quad \frac{\partial F}{\partial y} = N$$

  Y verificamos si se cumple la condición de exactitud:
  
  $$\frac{\partial M}{\partial y} = \frac{\partial N}{\partial x}$$

  <details>
   <summary><span style="color: #1073c5;">Ejemplo y procedimiento</span></summary>

   Considere la ecuación: $(2xy + 3) \, dx + (x^2 + 4y) \, dy = 0$
   1. Verificamos la condición de exactitud: $\frac{\partial M}{\partial y} = 2x \quad \text{y} \quad \frac{\partial N}{\partial x} = 2x$
   2. encontramos la función potencial: $\frac{\partial M}{\partial x} = 2xy + 3$ 
   3. Integrando con respecto a x: $F(x,y)=x^2y+3x+g(y)$
   4. Para encontrar g(y) encontramos la otra condición: $\frac{\partial N}{\partial y} = x^2+g(y) =x^2+4y$
   5. Igualando términos, obtenemos $g'(y)=4y$, lo que nos da $g(y)=2y^2$. Entonces, la solución general es:$$F(x, y) = x^2 y + 3x + 2y^2 = C $$

   </details>

  - __No exacta__: Si una ecuación de la forma $M(x, y) \, dx + N(x, y) \, dy = 0$ no cumple con la condición de exactitud, se dice que es no exacta.

<details>
  <summary><b>Ecuaciones diferenciales de Bernoulli</b></summary>

  Son un tipo específico de ecuaciones diferenciales de primer orden que tienen la forma:

  $$ y' + p(x)y = q(x)y^n $$

  <details>
   <summary><span style="color: #1073c5;">Procedimiento</span></summary>
  
   1. Reescribir la Ecuación en Forma Estándar: $ y' + p(x)y = q(x)y^n $
   2. Definir $y=u^{\frac{1}{1-n}}$
   3. Sustituyendo y multiplicando por $(1-n)$ tenemos: $$\frac{du}{dx} + (1-n)p(x)u = (1-n)q(x)$$ Esta es una ecuación diferencial lineal en u
   4. Utilizando el método estándar para resolver ecuaciones diferenciales lineales, encontramos el factor integrante $\mu$: $$ \mu= e^{\int (1-n)p(x)dx} $$
   5. Multiplicamos la ecuación por el factor integrante: $\mu(x) \frac{dv}{dx} + \mu(x)(1-n)p(x)u = \mu(x)(1-n)q(x)$
   6. Que se simplifica en: $\frac{d}{dx} [\mu(x)u] = \mu(x)(1-n)q(x)$
   7. Integrando ambos lados con respecto a $x$: $\mu(x)u = \int \mu(x)(1-n)q(x) \, dx + C$
   8. Resolvemos para u $$u = \frac{1}{\mu(x)} \left( \int \mu(x)(1-n)q(x) \, dx + C \right) $$
   9. Nos vevolvemos para la variable y y nos queda: $$y = \left( \frac{1}{\mu(x)} \left( \int \mu(x)(1-n)q(x) \, dx + C \right) \right)^{\frac{1}{1-n}} $$
  </details>
</details>

<details>
  <summary><b>Ecuación de Cauchy-Euler</b></summary>

  También conocida como ecuación de Euler-Cauchy, es una ecuación diferencial lineal de segundo orden que tiene la forma general: $$x^2 y'' + a x y' + b y = 0$$

  <details>
   <summary><span style="color: #1073c5;">Procedimiento</span></summary>

   1. Suponemos una solución de la forma: $y = x^r$, donde $r$ es una constante a determinar.
   2. Derivar la solución respecto a x: $$y' = r x^{r-1} \\ y'' = r (r-1) x^{r-2}$$
   3. Sustituir en la Ecuación Original: $$x^2 y'' + a x y' + b y = 0 \\ r (r-1) + a r + b = 0$$
   4. Resolver la Ecuación Característica: $r^2 + (a-1)r + b = 0$
   5. Dependiendo de las raices $r_1$ y $r_2$  de la ecuación característica, la solución general de la ecuación diferencial puede ser:
      1. Si las raices $r_1$ y $r_2$ son distintas de 0: $y = C_1 x^{r_1} + C_2 x^{r_2}$
      2. Si las raices $r_1$ y $r_2$ son reales e iguales: $ y = C_1 x^{r_1} + C_2 x^{r_1} \ln(x)$
      3. Si las raíces son complejas conjugadas $r$: $y = x^{\alpha} \left( C_1 \cos(\beta \ln(x)) + C_2 \sin(\beta \ln(x)) \right)$
</details>

### ☁️ Métodos de solución de EDO

<details>
  <summary><b>Separación de Variables</b></summary>

  **Descripción:**
  Este método se utiliza para resolver EDOs de primer orden que se pueden escribir en la forma $\frac{dy}{dx} = g(x)h(y)$. Consiste en separar las variables $x$ e $y$ y luego integrar ambos lados de la ecuación.

  <details>
   <summary><span style="color: #1073c5;">Ejemplo</span></summary>
  
   **Ejemplo**
   Considera la ecuación $\frac{dy}{dx} = xy$.
   
   1. **Separar las variables:**
      $$
      \frac{dy}{y} = x \, dx
      $$
   
   2. **Integrar ambos lados:**
      $$
      \int \frac{1}{y} \, dy = \int x \, dx
      $$
      $$
      \ln |y| = \frac{x^2}{2} + C
      $$
   
   3. **Resolver para $y$:**
      $$
      y = e^{\frac{x^2}{2} + C} = e^C e^{\frac{x^2}{2}}
      $$
      $$
      y = C' e^{\frac{x^2}{2}}
      $$
      donde $C' = e^C$ es una constante.
    
  </details>

</details>
   

<details>
  <summary><b>Integración por Factores</b></summary>

  **Descripción:**
  Este método se usa para resolver ecuaciones lineales de primer orden de la forma $\frac{dy}{dx} + p(x)y = q(x)$. Se multiplica la ecuación por un factor integrante para simplificar la solución.
  
  <details>
   <summary><span style="color: #1073c5;">Ejemplo</span></summary>
   
   Considera la ecuación $\frac{dy}{dx} + y = e^x$.
   
   1. **Identificar el factor integrante:** Para esta ecuación, el factor integrante $\mu(x)$ es:
      $$
      \mu(x) = e^{\int 1 \, dx} = e^x
      $$
   
   2. **Multiplicar ambos lados de la ecuación por el factor integrante:**
      $$
      e^x \frac{dy}{dx} + e^x y = e^x e^x
      $$
      $$
      e^x \frac{dy}{dx} + e^x y = e^{2x}
      $$
   
   3. **Simplificar la ecuación:**
      $$
      \frac{d}{dx} (e^x y) = e^{2x}
      $$
   
   4. **Integrar ambos lados:**
      $$
      e^x y = \int e^{2x} \, dx = \frac{e^{2x}}{2} + C
      $$
   
   5. **Resolver para $y$:**
      $$
      y = \frac{e^x}{2} + Ce^{-x}
      $$
   
   </details>
</details>

<details>
  <summary><b>Método de Sustitución</b></summary>

  **Descripción:**
  Este método se emplea para simplificar ecuaciones complejas, a menudo cambiando las variables o haciendo una sustitución adecuada.
  
  <details>
   <summary><span style="color: #1073c5;">Ejemplo</span></summary>

   Considera la ecuación $\frac{dy}{dx} = \frac{y}{x} + \frac{y^2}{x^2}$.
   
   1. **Sustitución:** Usa la sustitución $v = \frac{y}{x}$, así $y = vx$ y $\frac{dy}{dx} = v + x \frac{dv}{dx}$.
   
   2. **Sustituir en la ecuación:**
      $$
      v + x \frac{dv}{dx} = \frac{vx}{x} + \frac{(vx)^2}{x^2}
      $$
      $$
      v + x \frac{dv}{dx} = v + v^2
      $$
   
   3. **Simplificar:**
      $$
      x \frac{dv}{dx} = v^2
      $$
      $$
      \frac{dv}{v^2} = \frac{dx}{x}
      $$
   
   4. **Integrar ambos lados:**
      $$
      \int \frac{dv}{v^2} = \int \frac{dx}{x}
      $$
      $$
      -\frac{1}{v} = \ln |x| + C
      $$
      $$
      v = -\frac{1}{\ln |x| + C}
      $$
   
   5. **Deshacer la sustitución:**
      $$
      \frac{y}{x} = -\frac{1}{\ln |x| + C}
      $$
      $$
      y = -\frac{x}{\ln |x| + C}
      $$
   </details>
</details>

<details>
  <summary><b>Método de Variación de Parámetros</b></summary>

  **Descripción:**
  Se utiliza para resolver ecuaciones no homogéneas de segundo orden. Se supone una solución general de la forma $y = y_h + y_p$, donde $y_h$ es la solución homogénea y $y_p$ es una solución particular.
  
  <details>
   <summary><span style="color: #1073c5;">Ejemplo</span></summary>
  
   Considera la ecuación $\frac{d^2y}{dx^2} - 3 \frac{dy}{dx} + 2y = e^x$.
   
   1. **Resolver la ecuación homogénea:** $ \frac{d^2y}{dx^2} - 3 \frac{dy}{dx} + 2y = 0 $
      - Encontramos las raíces del polinomio característico $r^2 - 3r + 2 = 0$, que son $r = 1$ y $r = 2$.
      - La solución homogénea es:
        $$
        y_h = C_1 e^{2x} + C_2 e^x
        $$
   
   2. **Buscar una solución particular:** Supón $y_p = A e^x$. Sustituye en la ecuación no homogénea:
      $$
      \frac{d^2(A e^x)}{dx^2} - 3 \frac{d(A e^x)}{dx} + 2(A e^x) = e^x
      $$
      - Calcula las derivadas:
        $$
        \frac{d(A e^x)}{dx} = A e^x
        $$
        $$
        \frac{d^2(A e^x)}{dx^2} = A e^x
        $$
      - Sustituye en la ecuación:
        $$
        A e^x - 3A e^x + 2A e^x = e^x
        $$
        $$
        0 = e^x
        $$
      - Como esto no da una solución válida, prueba con $y_p = x e^x$:
        $$
        \frac{d^2(x e^x)}{dx^2} = e^x + 2x e^x
        $$
        $$
        \frac{d(x e^x)}{dx} = e^x + x e^x
        $$
        $$
        \frac{d^2(x e^x)}{dx^2} - 3 \frac{d(x e^x)}{dx} + 2(x e^x) = e^x
        $$
        $$
        (e^x + 2x e^x) - 3(e^x + x e^x) + 2x e^x = e^x
        $$
        $$
        e^x = e^x
        $$
      - Por lo tanto, $y_p = x e^x$ es una solución particular.
   
   3. **Combinar la solución homogénea y la particular:**
      $$
      y = C_1 e^{2x} + C_2 e^x + x e^x
      $$
   </details>
</details>

<details>
  <summary><b>Transformada de Laplace</b></summary> 

  **Descripción:**
  Convierte una EDO en una ecuación algebraica usando la transformada de Laplace, que facilita la solución. Después de encontrar la solución en el dominio de Laplace, se aplica la transformada inversa para obtener la solución en el dominio original.
  
  <details>
   <summary><span style="color: #1073c5;">Ejemplo</span></summary>

   Considera la ecuación $\frac{d^2y}{dx^2} + 3 \frac{dy}{dx} + 2y = 0$ con condiciones iniciales $y(0) = 1$ y $\frac{dy}{dx}(0) = 0$.
   
   1. **Aplicar la transformada de Laplace:**
      - La transformada de Laplace de $\frac{d^2y}{dx^2}$ es $s^2 Y(s) - s y(0) - \frac{dy}{dx}(0)$.
      - La transformada de Laplace de $\frac{dy}{dx}$ es $s Y(s) - y(0)$.
      - La transformada de Laplace de $y$ es $Y(s)$.
      - Aplicando la transformada de Laplace a la ecuación:
        $$
        \mathcal{L} \left\{ \frac{d^2y}{dx^2} \right\} + 3 \mathcal{L} \left\{ \frac{dy}{dx} \right\} + 2 \mathcal{L} \left\{ y \right\} = 0
        $$
        $$
        s^2 Y(s) - s y(0) - \frac{dy}{dx}(0) + 3 \left(s Y(s) - y(0)\right) + 2 Y(s) = 0
        $$
        $$
        s^2 Y(s) - s + 3s Y(s) - 3 + 2 Y(s) = 0
        $$
   
   2. **Simplificar y resolver para $Y(s)$:**
      $$
      (s^2 + 3s + 2) Y(s) = s + 3
      $$
      $$
      Y(s) = \frac{s + 3}{s^2 + 3s + 2}
      $$
   
      - Factoriza el denominador:
        $$
        s^2 + 3s + 2 = (s + 1)(s + 2)
        $$
      - Entonces:
        $$
        Y(s) = \frac{s + 3}{(s + 1)(s + 2)}
        $$
   
   3. **Aplicar la transformada inversa de Laplace:**
      - Usa fracciones parciales para simplificar:
        $$
        \frac{s + 3}{(s + 1)(s + 2)} = \frac{A}{s + 1} + \frac{B}{s + 2}
        $$
        - Encuentra $A$ y $B$:
          $$
          A(s + 2) + B(s + 1) = s + 3
          $$
          - Resolviendo para $A$ y $B$:
            $$
            A = 1, \quad B = 2
            $$
      - Entonces:
        $$
        Y(s) = \frac{1}{s + 1} + \frac{2}{s + 2}
        $$
   
      - La transformada inversa de Laplace es:
        $$
        y(x) = e^{-x} + 2 e^{-2x}
        $$
   </details>
</details>
   

<details>
  <summary><b>Series de potencias</b></summary>

  Las series de potencias son una herramienta poderosa para resolver ecuaciones diferenciales ordinarias (EDO) cuando otras técnicas no son adecuadas. La idea básica es suponer que la solución de la ecuación diferencial puede expresarse como una serie de potencias y luego determinar los coeficientes de esta serie.

  <details>
   <summary><span style = "color: #1073c5">Procedimiento</span></summary>

   1. Supongamos que la solución de la EDO tiene la forma de una serie de potencias: $ y(x) = \sum_{n=0}^{\infty} a_n x^n $
   2. Derivamos la serie término a término para obtener $y'$ y $y''$: $$y'(x) = \sum_{n=1}^{\infty} n a_n x^{n-1} \\ y''(x) = \sum_{n=2}^{\infty} n (n-1) a_n x^{n-2}$$
   3. Sustituimos la serie para $y$, $y'$ y $y''$: $$p(x) = \sum_{n=0}^{\infty} p_n x^n \\ q(x) = \sum_{n=0}^{\infty} q_n x^n$$
   4. Para simplificar, reindexamos las series y agrupamos términos similares. La idea es combinar todas las series en una sola serie y luego igualar los coeficientes de $x^{n} a 0 para cada $n$, Esto nos da una relación recursiva para los coeficientes $a_n$.
   5. Utilizamos la relación recursiva obtenida para calcular los coeficientes $a_n$.Estos coeficientes determinan la solución de la ecuación diferencial.

### ☁️ Aplicaciones de EDO

<details>
  <summary><b>Crecimiento poblacional</b></summary>

  Los problemas de crecimiento poblacional son un clásico ejemplo de aplicación de ecuaciones diferenciales ordinarias (EDO). Vamos a considerar dos modelos básicos: el modelo de crecimiento exponencial y el modelo de crecimiento logístico.

  1. Modelo de crecimiento exponencial

     La ecuación diferencial para el crecimiento exponencial es: $$ \frac{dP}{dt} = rP $$

     La solución es: $$P(t) = P_0 e^{rt}$$

  2. Crecimiento Logístico

      El crecimiento logístico es un modelo más realista del crecimiento poblacional que tiene en cuenta la capacidad limitada de los recursos. A diferencia del crecimiento exponencial, que asume un crecimiento infinito, el crecimiento logístico incorpora una capacidad de carga $K$ , que es el límite máximo de la población que el entorno puede soportar.

      La ecuación diferencial para el crecimiento logístico es: $$ \frac{dP}{dt} = rP \left( 1 - \frac{P}{K} \right) $$

      La solución es: $$P(t) = \frac{K}{1 + \left( \frac{K - P_0}{P_0} \right)e^{-rt}}$$

<details>
  <summary><b>Ley de enfriamiento de Newton</b></summary>

  La ecuación diferencial que describe la ley de enfriamiento de Newton es: $$\frac{dT}{dt} = -k (T - T_{\text{env}})$$

  La solución de esta ecuación es: $$T(t) = T_{\text{env}} + (T_0 - T_{\text{env}}) e^{-kt}$$

  

<details>
  <summary><b>Mezclas</b></summary>

  Estos problemas generalmente involucran la mezcla de soluciones o sustancias en un tanque donde las tasas de entrada y salida afectan la concentración de la sustancia en el tanque.

   Consideremos un tanque que contiene un volumen $V$ de líquido con una cierta concentración de una sustancia. Supongamos que una solución entra al tanque a una tasa de $r_{in}$ litros por minuto con una concentración de $C_{in}$ gramos por litro y que la mezcla en el tanque se mantiene uniforme. La mezcla sale del tanque a una tasa de $r_{out}$ litros por minuto. Queremos encontrar la concentración de la sustancia en el tanque en función del tiempo.

   __Planteamiento del Problema__

   Sea $C(t)$ la concentración de la sustancia en el tanque en el tiempo $t$. La tasa de cambio de la cantidad de sustancia en el tanque es la diferencia entre la tasa de entrada y la tasa de salida de la sustancia.

   **_Ecuación Diferencial:_** La cantidad de sustancia que entra al tanque por unidad de tiempo es $r_{in}\cdot C_{in}$, y la cantidad que sale por unidad de tiempo es $r_{out}\cdot C(t)$ La ecuación diferencial para la cantidad de sustancia $Q(t)$ en el tanque es:$$\frac{dC(t)}{dt} = \frac{r_{\text{in}}}{V} C_{\text{in}} - \frac{r_{\text{out}}}{V} C(t)$$

   **_La solución general de la ecuación es:_** $$C(t) = \frac{r_{\text{in}} C_{\text{in}}}{r_{\text{out}}} + \left( C_0 - \frac{r_{\text{in}} C_{\text{in}}}{r_{\text{out}}} \right) e^{-\frac{r_{\text{out}}}{V}t}$$ donde $C_0$ es la concentración inicial en $t=0$




\cdot
